```python
import pandas as pd
```

```python
df = pd.read_csv("ULTA.csv")
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.index = pd.DatetimeIndex(df['date']).floor('D')
df.index
```

```python
all_days = pd.date_range(df.index.min(), df.index.max(), freq='D')
df.reindex(all_days).to_csv('ULTAAllDates.csv')
```
